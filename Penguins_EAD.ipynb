```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('penguins_size.csv')
```

```python
data.head()
```

## Descrição do Dataset:

* Esse dataset contém dados coletados das características físicas e geográficas de três espécies de pinguins: Adelie, Chinstrap e Gentoo.

## Descrição dos atributos:

* species: Espécie do pinguin;
* island: A ilha que o pinguin habita;
* culmen_length_mm: O tamanho da crista superior do bico do pinguin;
* culmen_depth_mm: A altura da crista superior do bico do pinguin;
* flipper_length_mm: O tamanho da nadadeira do pinguin;
* body_mass_g: Massa corporal do pinguin;
* sex: Sexo do pinguin;

```python
data.describe()
```

De acordo com a média, mediana e desvio padrão, os dados encontram-se dentro do esperado e não possuem diferenças fora do normal.

```python
data.info()
```

# Deletando valores faltantes

```python
data.dropna(axis=0, inplace=True)
```

```python
data.info()
```

# Exploração dos Dados

## Pinguins por Espécies

```python
sns.countplot(x=data['species'])
```

Podemos visualizar que a espécie Chinstrap tem menos registros que as outras espécies.

## Pinguis por Ilha

```python
sns.countplot(x=data['island'])
```

Podemos verificar que existem menos pinguins na ilha Torgersen em relação as outras ilhas.

```python
sns.countplot(x=data['island'], hue=data['species'])
```

A espécie Adelie é encontrado nas 3 ilhas, a espécie Chinstrap é encontrada somente na ilha Dream e a espécie Gentoo somente na ilha Biscoe

## Pinguins por Sexo

```python
sns.countplot(x=data['sex'])
```

Foi encontrado um registro com o sexo preenchido com um ".". Esse registro será excluído.

```python
data = data[data['sex'] != '.']
```

```python
sns.countplot(x=data['sex'])
```

A população dos pinguins está balanceada em relação ao sexo.

```python
data.head()
```

## Medidas por Espécie

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12), constrained_layout=True)
sns.boxplot(x='species', y='culmen_length_mm', hue='sex', data=data, orient='v', ax=axs[0][0])
sns.boxplot(x='species', y='culmen_depth_mm', hue='sex', data=data, orient='v', ax=axs[0][1])
sns.boxplot(x='species', y='flipper_length_mm', hue='sex', data=data, orient='v', ax=axs[1][0])
sns.boxplot(x='species', y='body_mass_g', hue='sex', data=data, orient='v', ax=axs[1][1])
```

### Características por Espécie:

* Adelie possui crista superior do bico pequena e profunda, nadadeiras médias e massa corporal média;
* Chinstrap possui crista superior do bico grande e profunda, nadadeiras médias e a espécie de menor massa corporal entre as três;
* Gentoo possui crista superior do bico grande e de profundade pequena (menor entre as espécies), nadadeiras longas e a maior massa corporal em relação as outras espécies;
* As medidas dos penguins machos são maiores em todas as espécies;

## Medidas da espécie Adelie em diferentes Regiões

```python
data_adelie = data[data['species'] == 'Adelie']

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12), constrained_layout=True)
sns.boxplot(x='species', y='culmen_length_mm', hue='island', data=data_adelie, orient='v', ax=axs[0][0])
sns.boxplot(x='species', y='culmen_depth_mm', hue='island', data=data_adelie, orient='v', ax=axs[0][1])
sns.boxplot(x='species', y='flipper_length_mm', hue='island', data=data_adelie, orient='v', ax=axs[1][0])
sns.boxplot(x='species', y='body_mass_g', hue='island', data=data_adelie, orient='v', ax=axs[1][1])
```

Podemos notar que não há diferença significativa entre as medidas das espécies que vivem em diferentes regiões.

```python
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(12, 12), constrained_layout=True)
sns.scatterplot(x='culmen_length_mm', y='culmen_depth_mm', hue='species', data=data, ax=axs[0][0])
sns.scatterplot(x='culmen_length_mm', y='flipper_length_mm', hue='species', data=data, ax=axs[0][1])
sns.scatterplot(x='culmen_length_mm', y='body_mass_g', hue='species', data=data, ax=axs[1][0])
sns.scatterplot(x='culmen_depth_mm', y='flipper_length_mm', hue='species', data=data, ax=axs[1][1])
sns.scatterplot(x='culmen_depth_mm', y='body_mass_g', hue='species', data=data, ax=axs[2][0])
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', hue='species', data=data, ax=axs[2][1])
```

```python
sns.heatmap(data.corr(), annot=True)
```

Podemos concluir uma forte correlação entre a massa corporal e o tamanho da nadadeira, uma relação média positiva entre o tamanho da crista superior do bico e o tamanho da nadadeira, e uma relação média negativa entre a profundidade da crista superior do bico e a nadadeira. 

Ou seja, temos uma tendência a ter um aumento da massa corporal e do tamanho da crista superior do bico conforme o aumento do tamanho da nadadeira e a profundidade da crista superior do bico tende a diminuir conforme o tamanho da nadadeira aumenta, independente da espécie.
